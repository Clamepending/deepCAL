```python
import vamtoolbox as vam
import numpy as np
import os
import time
# %matplotlib qt

dimension = [1920, 1080]
OutputFolder = "D:/archive/members/mark/128x128/SpeedTestDump//"
InputFolder = "D:/archive/members/mark/RotatedCenteredSTLfiles/"
name = "/1/2020.stl"
OutputPrefix = "Target"
num_angles = 360
n_iter  = 50
d_h = 0.85
d_l = 0.6
filterType = 'hamming'
method = 'OSMO'
resolution = 128
```

```python
#Optional Cell for testing speed over all files. Comment out if making a projection for a single file.
files = []

# Iterate over each folder
STL_folder = os.listdir(InputFolder)

folder_path = os.path.join(InputFolder)
    
# # Normalize the folder path to ensure consistent forward slashes
folder_path = os.path.normpath(folder_path)
    
# List all files in the folder with the ".stl" extension
for file in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file)
        
    # Normalize the file path to ensure consistent forward slashes
    file_path = os.path.normpath(file_path)
        
    if file.endswith('.stl') and os.path.isfile(file_path):
        files.append(file_path)

print("Timing the " + (str)(len(files)) + " files:", files)

start_Time = time.time()
times = []
for f in files:
    print(f)
    fileStart_time = time.time()

    targetgeo = vam.geometry.TargetGeometry(stlfilename = f, resolution = resolution)
    angles = np.linspace(0, 360 - 360/num_angles, num_angles)
    proj_geo = vam.geometry.ProjectionGeometry(angles, ray_type='parallel', CUDA=True)
    optimizer_params = vam.optimize.Options(method=method, n_iter=n_iter, d_h = d_h, d_l = d_l, filter=filterType, verbose='plot')
    opt_sino, opt_recon, error = vam.optimize.optimize(targetgeo, proj_geo, optimizer_params)
    
    fileEnd_time = time.time()
    times.append(fileEnd_time-fileStart_time)
    print(f + " Time: " + str(times[len(times)-1]))
    totaltime = time.time()
    print("Current Time: "+ str(totaltime-start_Time))

end_Time = time.time()

print("Total Time: "+ str(end_Time-start_Time))
    
```
